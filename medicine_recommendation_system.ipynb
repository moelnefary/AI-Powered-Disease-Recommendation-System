```python
# import important libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from xgboost import XGBClassifier

from sklearn.metrics import accuracy_score ,confusion_matrix,ConfusionMatrixDisplay
from sklearn.preprocessing import StandardScaler ,LabelEncoder
from sklearn.model_selection import train_test_split

```

```python
# load data
df_train =pd.read_csv("Training.csv")

symp_severity=pd.read_csv("Symptom-severity.csv") # wieght of each symptom
symtoms=pd.read_csv("symtoms_df.csv")

description=pd.read_csv("description.csv")

diets=pd.read_csv("diets.csv")
medications=pd.read_csv("medications.csv")
precautions=pd.read_csv("precautions_df.csv")
workout=pd.read_csv("workout_df.csv")
```

## discovering data

```python
print(df_train.shape)
df_train.head()
```

```python
prognosis=df_train["prognosis"].unique()
print(f" number of disease can be diagnosed : {len(prognosis)}")
prognosis
```

```python
X=df_train.drop("prognosis",axis=1)
y=df_train["prognosis"]
y.head()
```

```python
encoder=LabelEncoder()
y=encoder.fit_transform(y)
lable_to_disease=dict(enumerate(encoder.classes_))
disease_labels={d:l for l,d in lable_to_disease.items()}
```

```python
disease_labels
```

```python
print("X shape :" ,X.shape)
print("y shape :",y.shape)
```

```python
# splitting the dataset to train and test
x_train,x_test,y_train,y_test= train_test_split(X,y,test_size=.2,random_state=42)
```

```python
print(x_train.shape); print(x_test.shape)
```

```python
models={

        "logistc_regression" : LogisticRegression(random_state=42),
        "xgboost" :XGBClassifier(n_estimators=500 ,max_depth=6,learning_rate=.1,subsample=.8,eval_metric="mlogloss"),
        "random_forest" : RandomForestClassifier(n_estimators=500,max_depth=None,min_samples_split=2 ,random_state=42),
        "svc":SVC(kernel="rbf",C=1,probability=True)
}
accuracies={}
conf_matrix={}
for name ,model in models.items():
  model.fit(x_train,y_train)
  y_pred=model.predict(x_test)
  accuracies[name]=accuracy_score(y_test,y_pred)
  conf_matrix[name]=confusion_matrix(y_test,y_pred)




```

```python
for n ,a in accuracies.items():
  print(f"accuracy of {n} is {a} \n")
```

```python
xgb_model=models["xgboost"]
xgb_model.predict(x_test.iloc[[0]])
print(xgb_model.predict(x_test.iloc[[0]]))
print(y_test[0])
```
